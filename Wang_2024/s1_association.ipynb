```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("processed_data.csv", index_col=0)
meta = pd.read_csv("data_from_RDS/cell_metadata.csv", index_col=0)

df
```

```python
#meta.index
#df.index
meta.columns
```

```python
df["response"] = meta.loc[df.index, "response"]
#df["treatment"] = meta.loc[df.index, "treatment"]
df["timepoint"] = meta.loc[df.index, "timepoint"]
df["patient_alias"] =meta.loc[df.index, "patient_alias"]
df
#print(df["response"].unique())
```

```python
patients_df = (
    df.groupby(["patient_alias", "response", "timepoint"])
      .agg(
          total_cells=("cancer_reactive", "size"),
          cancer_reactive_cells=("cancer_reactive", "sum")
      )
      .reset_index()
)
patients_df
```

```python
# filter to baseline only
baseline_df = patients_df[patients_df["timepoint"] == "Baseline"]

# make boxplot
plt.figure(figsize=(6,4))
baseline_df.boxplot(
    column="cancer_reactive_cells",
    by="response",
    grid=False
)

plt.suptitle("")  # remove automatic title
plt.title("Baseline: Cancer-reactive cells per patient by response")
plt.xlabel("Response type")
plt.ylabel("# cancer-reactive cells per patient")
plt.ylim(bottom=0)
plt.show()
```

```python
groups={"Good response":['CR', 'PR'], "Poor response":['PD', 'SD']}
for group in groups:
    sub_df = df[df["response"].isin(groups[group])]
    # count number of cancer reactive cells per timepoint
    counts = (
        sub_df[sub_df["cancer_reactive"] == True]
        .groupby("timepoint")
        .size()
        .reset_index(name="cancer_reactive_count")
    )

    # sort by timepoint if they’re categorical
    time_order = ["Baseline", "Follow Up 1", "Follow Up 2", "Follow Up 3", "Follow Up 4", "Follow Up 5", "Follow Up 6"]
    counts["timepoint"] = pd.Categorical(counts["timepoint"], categories=time_order, ordered=True)
    counts = counts.sort_values("timepoint")

    # plot
    plt.figure(figsize=(8,5))
    plt.plot(counts["timepoint"], counts["cancer_reactive_count"], marker="o")
    plt.xlabel("Timepoint")
    plt.ylabel("Number of cancer reactive cells")
    plt.title("Cancer reactive cells across timepoints - " + group)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
