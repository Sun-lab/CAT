```python
df["response"] = meta.loc[df.index, "response"]
# df["treatment"] = meta.loc[df.index, "treatment"]
df["timepoint"] = meta.loc[df.index, "timepoint"]
df["patient_alias"] =meta.loc[df.index, "patient_alias"]
df
#print(df["response"].unique())
```

```python
patients_df = (
    df.groupby(["patient_alias", "response", "timepoint"])
      .agg(
          total_cells=("cancer_reactive_per_cell_CD8", "size"),
          cancer_reactive_cells=("cancer_reactive_per_cell_CD8", "sum")
      )
      .reset_index()
)
print(patients_df.head())
```

```python
# filter to baseline only
baseline_df = patients_df[patients_df["timepoint"] == "Baseline"]

# make boxplot
plt.figure(figsize=(6,4))
baseline_df.boxplot(
    column="cancer_reactive_cells",
    by="response",
    grid=False
)

plt.suptitle("")  # remove automatic title
plt.title("Baseline: Cancer-reactive cells per patient by response")
plt.xlabel("Response type")
plt.ylabel("# cancer-reactive cells per patient")
plt.ylim(bottom=0)
plt.show()
```

```python
# Count samples per response × timepoint
counts = (
    df.groupby(["timepoint", "response"])
      .size()
      .reset_index(name="n_samples")
)

print(counts)
```

```python
# Create boxplot faceted by timepoint
df = patients_df[~patients_df["timepoint"].isin(["Follow Up 5", "Follow Up 6"])].copy()
resp_order = ["PD", "SD", "PR", "CR"]
df["response"] = pd.Categorical(df["response"], categories=resp_order, ordered=True)

df["prop_cancer_reactive"] = df["cancer_reactive_cells"] / df["total_cells"]
import seaborn as sns

# Boxplots (hollow, no fliers)
g = sns.catplot(
    data=df,
    x="response", y="prop_cancer_reactive",
    col="timepoint",
    kind="box",
    order=resp_order,
    col_wrap=3,
    sharey=True,
    height=4, aspect=0.85,
    showcaps=True,
    boxprops={"facecolor": "none", "edgecolor": "black"},
    whiskerprops={"color": "black"},
    medianprops={"color": "red", "linewidth": 1.5},
    capprops={"color": "black"},
    showfliers=False    # remove outlier dots
)

# Overlay individual points (larger)
g.map_dataframe(
    sns.stripplot,
    x="response", y="prop_cancer_reactive",
    order=resp_order,
    jitter=True,
    alpha=0.6,
    size=5,          # larger points
    color="black",
    zorder=1
)

g.set_axis_labels("Response", "Proportion cancer reactive cells")
g.set_titles("{col_name}")
for ax in g.axes.flatten():
    ax.grid(True, axis="y", alpha=0.25)

plt.tight_layout()
plt.show()

```
