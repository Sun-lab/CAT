```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

pos_sigs_CD8 = [
    "CD8Lowery_pos_243g", 
    "CD8Oliveira_TTE_100g", 
    "CD8Oliveira_pos_74g", 
    "CD8Yost_CD8_Exh_100g", 
    "ave_Hanada_pos_27g"
]
neg_sigs_CD8 = [
    #"CD8Lowery_neg_99g",
    "ave_Hanada_neg_5g", 
    "ave_Oliveira_virus_26g",
]

pos_sigs_CD4 = [
    "CD4Hanada_pos_9g", 
    "CD4Caushi_Tfh2_66g", 
    "CD4Jansen_TermDiff_73g", 
    "CD4Lowery_pos_40g", 
    "CD4Oh_CXCL13_50g"
]
neg_sigs_CD4 = [
    "CD4Lowery_neg_37g", 
    "CD4Hanada_neg_4g"
]


t_cells = {'CD4':[
'c01_CD4_Tn_CCR7',
 'c02_CD4_Tn_SELL',
 'c03_CD4_Tn_NR4A2',
 'c04_CD4_Tcm_ANXA1',
 'c05_CD4_Tcm_GPR183',
 'c06_CD4_Trm_HSPA1A',
 'c07_CD4_Th17_CTSH',
 'c08_CD4_Tfh_CXCL13_IL6ST',
 'c09_CD4_Th1_CXCL13_HAVCR2',
 'c10_CD4_Temra_GZMB',
 'c11_CD4_Treg_FOXP3',
 'c12_CD4_Treg_KLRB1',
 'c13_CD4_Treg_TNFRSF9',
 'c14_CD4_MT'], 'CD8':[
 'c15_CD8_Tn_CCR7',
 'c16_CD8_Tn_SELL',
 'c17_CD8_Tcm_GPR183',
 'c18_CD8_Tcm_ANXA1',
 'c19_CD8_Tem_CMC1',
 'c20_CD8_Tem_GZMK',
 'c21_CD8_Trm_XCL1',
 'c22_CD8_Trm_HSPA1B',
 'c23_CD8_Tex_LAYN',
 'c24_CD8_Temra_CX3CR1',
 'c25_CD8_Temra_TYROBP',
 'c26_CD8_MAIT_KLRB1',
 'c27_CD8_MAIT_SLC4A10',
 'c28_CD8_IEL_CD160']
}
def classify_cells(df, t_cells, cell_type, pos_sigs, neg_sigs):
    sub = df[df["SubCellType"].isin(t_cells[cell_type])].copy()
    cols = pos_sigs + neg_sigs

    # z‐score the signatures _within_ this subset
    sub[cols] = sub[cols].apply(zscore)

    # compute mean “pos” and “neg” scores
    sub[f"pos_score_{cell_type}"] = sub[pos_sigs].mean(axis=1)
    sub[f"neg_score_{cell_type}"] = sub[neg_sigs].mean(axis=1)

    # pick cutoffs (here: median)
    pos_cut = sub[f"pos_score_{cell_type}"].median()
    neg_cut = sub[f"neg_score_{cell_type}"].median()

    # classify per cell
    sub[f"cancer_reactive_per_cell_{cell_type}"] = (
        (sub[f"pos_score_{cell_type}"] >  pos_cut) &
        (sub[f"neg_score_{cell_type}"] <  neg_cut)
    )

    return sub

# run it for each cell‐type
cd8_called = classify_cells(df, t_cells, "CD8", pos_sigs_CD8, neg_sigs_CD8)
cd4_called = classify_cells(df, t_cells, "CD4", pos_sigs_CD4, neg_sigs_CD4)

df_combined = pd.concat([cd8_called, cd4_called]).sort_index()
df_combined
```

# Clustering based on signature genes

```python
import pickle
with open("signatures_CD8.pkl", "rb") as f:
    sigs_CD8 = pickle.load(f)

print({k: len(v) for k, v in sigs_CD8.items()})

with open("signatures_CD4.pkl", "rb") as f:
    sigs_CD4 = pickle.load(f)

print({k: len(v) for k, v in sigs_CD4.items()})
```

```python
import pandas as pd
import scanpy as sc
import numpy as np
features = pd.read_csv("GSE236581_features.tsv.gz", compression='gzip', sep='\t', header=None)

barcodes = pd.read_csv("GSE236581_barcodes.tsv.gz", compression='gzip', header=None)

from scipy.io import mmread

import gzip
with gzip.open("GSE236581_counts.mtx.gz", 'rb') as f:
    counts = mmread(f).tocsc()  # or .tocsr()
print(counts.shape)

import anndata
features.columns = ["gene_id", "gene_name", "feature_type"]
barcodes.columns = ["barcode"]

counts = counts.transpose()  

# Create the AnnData object
adata = anndata.AnnData(X=counts)

# AnnData “obs” (rows) will correspond to cells
adata.obs_names = barcodes["barcode"].values

# AnnData “var” (columns) will correspond to genes
adata.var_names = features["gene_id"].values

# Store additional gene annotations
adata.var["gene_name"] = features["gene_name"].values
adata.var["feature_type"] = features["feature_type"].values

adata.obs = adata.obs.join(metadata)
adata_all = adata[adata.obs["SubCellType"].isin(t_cell_types)]
```

```python
all_genes_flat_CD8 = [item for sublist in sigs_CD8.values() for item in sublist]
all_genes_flat_CD4 = [item for sublist in sigs_CD4.values() for item in sublist]
subset_genes_CD8 = list(set(all_genes_flat_CD8))
subset_genes_CD4 = list(set(all_genes_flat_CD4))
```

```python
matched_genes_CD8 = []
matched_genes_CD4 = []

for s1 in subset_genes_CD8:
    if s1 in list(adata_all.var['gene_name']):
        matched_genes_CD8.append(s1)
for s1 in subset_genes_CD4:
    if s1 in list(adata_all.var['gene_name']):
        matched_genes_CD4.append(s1)    
# Print match info
print(f" {len(matched_genes_CD4)}/{len(subset_genes_CD4)} CD4 genes are found.")
print(f" {len(matched_genes_CD8)}/{len(subset_genes_CD8)} CD8 genes are found.")
```

```python
adata = adata_all.copy()
```

```python
%matplotlib inline
def run_cluster(sub_adata, key: str):
    sc.pp.scale(sub_adata)
    sc.tl.pca(sub_adata, svd_solver="arpack")
    sc.pp.neighbors(sub_adata, n_neighbors=15, n_pcs=25)
    sc.tl.umap(sub_adata)
    sc.tl.leiden(sub_adata, resolution=0.3)
    sc.pl.umap(sub_adata, color="leiden", title=key)
    return sub_adata.obs.rename(columns={"leiden": f"cluster_{key}"})

cd4_mask = adata.obs["SubCellType"].isin(t_cells['CD4'])
cd8_mask = adata.obs["SubCellType"].isin(t_cells['CD8'])


leiden_cd4 = run_cluster(adata[cd4_mask, matched_genes_CD4].copy(), "CD4")
leiden_cd8 = run_cluster(adata[cd8_mask, matched_genes_CD8].copy(), "CD8")

clusters_wide = pd.concat([leiden_cd4, leiden_cd8], sort=False)
```

```python
merged_df = df_combined.join(clusters_wide[['cluster_CD4','cluster_CD8']])
```

```python
merged_df["cluster"] = merged_df.apply(
    lambda r: r["cluster_CD4"] if r["SubCellType"] in t_cells['CD4']
              else r["cluster_CD8"],
    axis=1
)

merged_df["cancer_reactive_per_cell"] = merged_df.apply(
    lambda r: r["cancer_reactive_per_cell_CD4"] if r["SubCellType"] in t_cells['CD4']
              else r["cancer_reactive_per_cell_CD8"],
    axis=1
)
is_cd4 = merged_df["SubCellType"].isin(t_cells['CD4'])

merged_df["cell_type"] = np.where(is_cd4, "CD4", "CD8")

# Group by both cell_type & cluster and compute the fraction reactive
cluster_frac = (
    merged_df
    .groupby(["cell_type", "cluster"])["cancer_reactive_per_cell"]
    .mean()
    .rename("reactive_fraction")
)

# if you want to inspect it as a nice DataFrame:
cluster_frac_df = cluster_frac.reset_index()
cluster_frac_df
```

```python
#  threshold and collect all (cell_type, cluster) pairs above it
threshold = 0.4
reactive_pairs = set(
    cluster_frac[cluster_frac > threshold].index
)

reactive_pairs
```

```python
# map back to every cell
merged_df["cancer_reactive_by_cluster"] = merged_df.apply(
    lambda r: True
              if (r["cell_type"], r["cluster"]) in reactive_pairs
              else False,
    axis=1
)
merged_df
```

```python
import pandas as pd
cell_data = pd.read_csv('cell_meta_data.csv', index_col = 0)
cell_data
```

```python
import pandas as pd
import numpy as np

# ------------------------------------------------------------------
# ---  short aliases to match dataframe -------------------
# ------------------------------------------------------------------
A_V   = 'V_gene(Alpha1)'
A_J   = 'J_gene(Alpha1)'
A_CDR = 'CDR3(Alpha1)'
B_V   = 'V_gene(Beta1)'
B_J   = 'J_gene(Beta1)'
B_CDR = 'CDR3(Beta1)'

CELLTYPE_COL      = 'cell_type'                # 'CD4' / 'CD8' labels
PREDICTED_COL     = 'SubCellType'       #  column holds label
SAMPLE = 'Patient'
# ------------------------------------------------------------------
#   ---  aggregation function ------------------------------------
# ------------------------------------------------------------------
def summarise_clonotype(df: pd.DataFrame) -> pd.Series:
    cd8 = df[df[CELLTYPE_COL] == 'CD8']
    cd4 = df[df[CELLTYPE_COL] == 'CD4']


    counts_per_sample = df.groupby(SAMPLE).size()
    abundance_per_sample = ','.join(counts_per_sample.astype(str))

    return pd.Series({
        'alpha_v_gene' : df[A_V].iloc[0],
        'alpha_j_gene' : df[A_J].iloc[0],
        'alpha_cdr3'   : df[A_CDR].iloc[0],
        'beta_v_gene'  : df[B_V].iloc[0],
        'beta_j_gene'  : df[B_J].iloc[0],
        'beta_cdr3'    : df[B_CDR].iloc[0],

        # --- mean signature scores directly from the four columns ---
        'mean_pos_score_CD8' : cd8['pos_score_CD8'].mean(),
        'mean_neg_score_CD8' : cd8['neg_score_CD8'].mean(),
        'mean_pos_score_CD4' : cd4['pos_score_CD4'].mean(),
        'mean_neg_score_CD4' : cd4['neg_score_CD4'].mean(),

        # --- concatenate every predicted label present in the group ---
        'predicted_celltype' : ','.join(df[PREDICTED_COL].astype(str).unique()),
        'SAMPLE' : ','.join(df[SAMPLE].astype(str).unique()),
        
        'abundance'     : len(df),
        'abundance_CD8' : len(cd8),
        'abundance_per_sample' : abundance_per_sample 
    })
# ------------------------------------------------------------------
#   ---  run the group-by ----------------------------------------
# ------------------------------------------------------------------
clonotype_cols = [A_V, A_J, A_CDR, B_V, B_J, B_CDR]   # unique identifier
tcr_df = (
    cell_data
      .groupby(clonotype_cols, sort=False, group_keys=False)
      .apply(summarise_clonotype)
      .reset_index(drop=True)
)
tcr_df
```

```python
tcr_df['num_samples'] = tcr_df['SAMPLE'].apply(lambda x: len(str(x).split(',')) if pd.notna(x) else 0)

# Sort the dataframe by number of samples (descending)
tcr_df_sorted = tcr_df.sort_values(by=['num_samples', 'abundance'], ascending=[False, False])
```
