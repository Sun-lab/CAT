```python
import pandas as pd
pd.set_option('display.max_columns', None)
out_CD8 = pd.read_csv("cell_meta_data_CD8.csv", index_col=[0, 1])
out_CD8
```

```python
out_CD8 = out_CD8.reset_index()
df_cell_only = out_CD8.sort_values(by=["cell_id", "TCR_entry_number"]).drop_duplicates(subset="cell_id", keep="first")
df_cell_only
```

```python
df_cell_only["cancer_reactive"] = (
    df_cell_only["cancer_reactive_per_cell"] 
  & df_cell_only["cancer_reactive_by_cluster"]
)
```

```python
df_prop = (
    df_cell_only
    .groupby(['LibraryName','Timepoint'])['cancer_reactive']
    .mean()                        # True/False are treated as 1/0 so mean gives the fraction
    .reset_index(name='prop_reactive')
)

df_prop
```

```python
quantiles = (
    df_prop
    .groupby('Timepoint')['prop_reactive']
    .quantile([0.25, 0.5, 0.75])    # 25th, 50th, 75th
    .unstack()
    .rename(columns={0.25:'Q1', 0.5:'Median', 0.75:'Q3'})
    .reset_index()
)
quantiles
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu     # non-parametric, unpaired

plt.figure(figsize=(6,4))
sns.boxplot(
    data=df_prop,
    x='Timepoint',
    y='prop_reactive',
    whis=[0,100],            
    showfliers=False        
)
sns.stripplot(
    data=df_prop,
    x='Timepoint',
    y='prop_reactive',
    color=".3",              # light gray points for each library
    jitter=True,
    size=4
)
plt.ylabel('Proportion cancer-reactive')
plt.xlabel('Timepoint')
plt.title('% Cancer-Reactive CD8 by Time')
plt.show()
```

# Plots of interest

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# â”€â”€â”€ (Re-)compute the fraction of cancer_reactive cells per Library Ã— Timepoint Ã— PFS_6M â”€â”€â”€
df_prop = (
    df_cell_only
    .groupby(['LibraryName', 'PFS_6M', 'Timepoint'])['cancer_reactive']
    .mean()                 # True/False â†’ fraction True
    .reset_index(name='prop_reactive')
)

# map 0/1 â†’ NR/RE
df_prop['Responder'] = df_prop['PFS_6M'].map({0: 'NR', 1: 'RE'})


# â”€â”€â”€ Create a single boxplot with hue='Responder' â”€â”€â”€
plt.figure(figsize=(6, 5))
ax = sns.boxplot(
    data=df_prop,
    x='Timepoint',
    y='prop_reactive',
    hue='Responder',
    palette='pastel',
    width=0.6,
    fliersize=0     # hide the default outlier points (weâ€™ll overlay them ourselves)
)

# â”€â”€â”€ Overlay the individual perâ€library points via a jittered stripplot â”€â”€â”€
sns.stripplot(
    data=df_prop,
    x='Timepoint',
    y='prop_reactive',
    hue='Responder',
    dodge=True,         # slide NR/RE points left/right within each Timepoint
    color='.3',         # dark gray for points
    size=4,
    jitter=True,
    alpha=0.7,
    ax=ax,
    linewidth=0
)

pvals = {}                               # {time-point: p-value}
for tp, grp in df_prop.groupby('Timepoint'):
    nr = grp.loc[grp['Responder'] == 'NR', 'prop_reactive']
    re = grp.loc[grp['Responder'] == 'RE', 'prop_reactive']
    _, p = mannwhitneyu(nr, re, alternative='two-sided')
    pvals[tp] = p

# â”€â”€ANNOTATE THE p-VALUES ON THE FIGURE â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# where to draw the bars / text
y_max = df_prop['prop_reactive'].max()
extra = (ax.get_ylim()[1] - ax.get_ylim()[0]) * 0.06   # vertical padding

for tp, p in pvals.items():
    x_center = tp              # the mid-point between the two boxes
    # horizontal bar above the two distributions
    ax.plot([x_center - .18, x_center + .18],
            [y_max + extra,  y_max + extra], lw=1.2, c='k')
    # text label
    ax.text(x_center, y_max + extra*1.25,
            f"p = {p:.3g}", ha='center', va='bottom', fontsize=10)



# â”€â”€â”€ Remove the duplicate legend from stripplot â”€â”€â”€
# After calling stripplot, seaborn will have added a second legend. We'll keep only one.
handles, labels = ax.get_legend_handles_labels()
# We know the first two handles/labels correspond to boxplot (NR, RE).
# The next two are from stripplot (NR, RE) â€“ so drop them.
ax.legend(handles=handles[:2], labels=labels[:2], title='Responder', bbox_to_anchor=(1.02, 1),loc='upper left')
#ax.legend(title='Responder', bbox_to_anchor=(1.02, 1), loc='upper left')

# â”€â”€â”€ Final formatting â”€â”€â”€
ax.set_ylim(top=y_max + extra*2)  
ax.set_xlabel('Timepoint')
ax.set_ylabel('Proportion cancerâ€reactive')
ax.set_title('% Cancerâ€Reactive CD8: NR vs RE (all timepoints)')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# for each library Ã— timepoint Ã— PFS_6M, get the fraction of cancer_reactive cells
df_prop = (
    df_cell_only
    .groupby(['LibraryName','PFS_6M','Timepoint'])['cancer_reactive']
    .mean()                            # mean of bools â†’ fraction True
    .reset_index(name='prop_reactive')
)

# map 0/1 labels
df_prop['Responder'] = df_prop['PFS_6M'].map({0:'NR', 1:'RE'})

g = sns.catplot(
    data=df_prop,
    col='Responder',
    x='Timepoint',
    y='prop_reactive',
    kind='box',
    height=4,
    aspect=1
)
# add per-library points on top
g.map_dataframe(
    sns.stripplot,
    x='Timepoint', y='prop_reactive',
    color='.3',
    jitter=True,
    size=4
)
g.set_axis_labels('Timepoint','Proportion cancer-reactive')
g.set_titles('{col_name}')
g.fig.suptitle('% Cancer-Reactive CD8: NR vs RE over Time', y=1.02)
plt.tight_layout()
plt.show()
```
